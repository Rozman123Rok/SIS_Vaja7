```python
#!/usr/bin/env python
# coding: utf-7

# In[94]:


# Rok Rozman
import numpy as np
#import matplotlib.pyplot as plt

#from matplotlib import animation
#from IPython.display import HTML

#import sounddevice as sd
from scipy.io import wavfile

#from scipy import signal


# In[95]:


DTMF_tabela = {(697, 1209): "1", (697, 1336): "2", (697, 1477): "3", (770, 1209): "4", (770, 1336): "5", (770, 1477): "6", (852, 1209): "7", (852, 1336): "8", (852, 1477): "9", (941, 1209): "*", (941, 1336): "0", (941, 1477): "#", (697, 1633): "A", (770, 1633): "B", (852, 1633): "C", (941, 1633): "D"}


# In[96]:


"""
Fvz, sig = wavfile.read('test.wav')
T = 1024
N = 1024

f, t, TFD = signal.spectrogram(sig, Fvz, window=('tukey', N/T), nperseg=N, noverlap=None)

plt.figure(figsize=(16, 12))
plt.pcolormesh(t, f, TFD)
plt.ylabel('frekvenca (Hz)')
plt.xlabel('훾as (s)')
plt.title('T = ' + str(T) + ', N = ' + str(N))
plt.show()
"""


# In[97]:


def analiziraj_dtmf(signal, vzorcevalna_frekvenca, min_cas_ton, min_cas_pavza):
    #print("signal: ", signal)
    #print("frek: ", vzorcevalna_frekvenca)
    
    izpis = ""
    korak = 1000
    
    for i in range(0, len(signal)-korak, korak):
        s = signal[i:i+korak]
        fourier = np.fft.fft(s)
        frekvenca = np.fft.fftfreq(s.size, d=1/vzorcevalna_frekvenca)
        
        # Iscemo najblizjo nizko freq
        spodna = np.where(frekvenca > 0)[0][0]
        zgorna = np.where(frekvenca > 1050)[0][0]

        freq = frekvenca[spodna:zgorna] # iscemo znotraj med 0 in 1050
        amp = abs(fourier.real[spodna:zgorna])

        nizka = freq[np.where(amp == max(amp))[0][0]]

        delta = 10 # kolko +/- je lahko
        najboljsa = 0 # kera je najbolj natancna si shranimo

        for f in [697, 770, 852, 941] : # gremo skozi vse nizke frekvence
            if abs(nizka-f) < delta : # ce je znotraj delte si jo shranimo
                delta = abs(nizka-f) # spremenimo delto
                najboljsa = f # si jo shranimo kot najboljso
        nizka = najboljsa # nasa nizka frek 
        
        # Iscemo najblizjo visoko freq
        spodna = np.where(frekvenca > 1100)[0][0]
        zgorna = np.where(frekvenca > 2000)[0][0]

        freq = frekvenca[spodna:zgorna] # iscemo samo frek med 1100 in 2000 
        amp = abs(fourier.real[spodna:zgorna])

        visoka = freq[np.where(amp == max(amp))[0][0]]

        delta = 10 # kolko +/- je lahko
        najboljsa = 0 # kera je najbolj natancna si shranimo

        for f in [1209, 1336, 1477, 1633] : # gledamo samo tote visoke frekvence
            if abs(visoka-f) < delta : # ce pase v delta
                delta = abs(visoka-f) # posodobimo delto
                najboljsa = f # si shranimo najboljso
        
        visoka = najboljsa # nasa visoka
        
        if nizka == 0 or visoka == 0 : # ce sta prazni torej ni pravih frek blo 
            izpis = ""
        elif DTMF_tabela[(nizka,visoka)] != izpis: # ce se se ta stevilka ni izpisala
            izpis = DTMF_tabela[(nizka,visoka)] # si jo damo na izpis
            print(izpis, end='') # jo izpisemo (end da nena naredi novi line)


# In[98]:


if __name__ == '__main__':
    print("Modul za DTMF analozo!")
    # pozenem test 
    # test.wav je 훾ist signal, vrednosti 123456789*00, minimalen 훾as tona 0.1 ms, minimalen 훾as pavze 0.1 s
    #a = np.load('primer_ki_ga_uporablja_test_s/s.npy')
    #print(a)
    #Fvz, sig = wavfile.read('test.wav')
    #Fvz, sig = wavfile.read('cisti_dtmf_123456789_0__min_pulse_0.1_min_pause_0.1.wav')
    # prekratki_dtmf_124679_min_pulse_0.1_min_pause_0.1.wav
    # malo_suma_dtmf_123_min_pulse_0.2_min_pause_0.1_noise_low.wav
    # srednje_suma_dtmf_123_min_pulse_0.2_min_pause_0.1_noise_med.wav
    # veliko_suma_dtmf_123_min_pulse_0.2_min_pause_0.1_noise_high.wav
    Fvz, sig = wavfile.read('cisti_dtmf_123456789_0__min_pulse_0.1_min_pause_0.1.wav')
    analiziraj_dtmf(sig,Fvz, 0.1, 0.1)

    
# In[ ]:




```
